##Libraries

```python
import pandas as pd
import numpy as np
import os
```

##Load Datasets

```python
def load_red_list_datasets(folder_name):
  datasets_dict = {}
  folder = os.listdir(folder_name)
  for file in folder:
    file_path = folder_name + "/" + file
    dataset_name = file[0:-4]
    datasets_dict[dataset_name] = pd.read_csv(file_path)
  return datasets_dict
```

```python
latam_folder = datasets_path + 'RedList_Latam'
latam = load_red_list_datasets(latam_folder)
```

```python
common_names = pd.read_csv(latam_folder + '/common_names_combined.csv')
```

##Analisis

lista de paises incluidos en al analisis

```python
south_america_list_filter = ['Argentina', 'Bolivia, Plurinational States of', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Falkland Islands (Malvinas)', 'French Guiana', 'Guyana', 'Paraguay','Peru', 'Suriname', 'Uruguay', 'Venezuela, Bolivarian Republic of']
```

filtrado y cambio de nombre

```python
south_america = latam['countries'][latam['countries']['name'].isin(south_america_list_filter)]
```

```python
south_america = south_america.rename(columns={'name' : "country_name"})
```

```python
rename_dict = {'Bolivia, Plurinational States of' : "Bolivia", 'Venezuela, Bolivarian Republic of': 'Venezuela', 'Falkland Islands (Malvinas)' : 'Malvinas'}
south_america = south_america.replace(rename_dict)
```

```python
south_america
```

tipo de peligro en el que estan las especies

```python
latam['assessments']['redlistCategory'].unique()
```

dataframe con todas las categorias de riesgo de cada animal

```python
animals_category = latam['assessments'][['internalTaxonId','redlistCategory']]
```

dataframe con los animales en riesgo

```python
threatened_animals = latam['assessments'][['internalTaxonId','redlistCategory']][~(latam['assessments']['redlistCategory'].isin(['Near Threatened', 'Data Deficient', 'Extinct', 'Extinct in the Wild']))]
```

dataframe con los animales que no estan en riesgo

```python
safe_animals = latam['assessments'][['internalTaxonId','redlistCategory']][(latam['assessments']['redlistCategory'].isin(['Near Threatened', 'Data Deficient']))]
```

dataframe con todos los animales no extintos

```python
all_animals_not_extinct = latam['assessments'][['internalTaxonId','redlistCategory']][~(latam['assessments']['redlistCategory'].isin(['Extinct', 'Extinct in the Wild']))]
```

```python
threatened_animals
```

merge entre animales de sur america y su tipo de riesgo

todos los animales, extintos incluidos



```python
south_america_animals = pd.merge(south_america, animals_category, how='inner')
```

todos los animales, no extintos

```python
south_america_animals_not_extinct = pd.merge(south_america, all_animals_not_extinct, how='inner')
```

animales fuera de peligro

```python
south_america_not_threatened_animals = pd.merge(south_america, safe_animals, how='inner')
```

animales en peligro

```python
south_america_threatened_animals = pd.merge(south_america, threatened_animals, how='inner')
```

```python
south_america_threatened_animals
```

agrupamiento por pais todos los animales no extintos

```python
number_of_south_america_bot_extinct_species_by_country = south_america_animals_not_extinct.groupby('country_name')['internalTaxonId'].count().reset_index().rename(columns={'internalTaxonId': 'Cantidad'})
```

```python
number_of_south_america_bot_extinct_species_by_country
```

```python
number_of_south_america_bot_extinct_species_by_country.to_csv(custom_datasets_path + 'number_of_south_america_bot_extinct_species_by_country.csv')
```

agrupamiento por pais animales a salvo

```python
number_of_south_america_safe_species = south_america_not_threatened_animals.groupby('country_name')['internalTaxonId'].count().reset_index().rename(columns={'internalTaxonId': 'Cantidad'})
number_of_south_america_safe_species['in_risk'] = 'true'
number_of_south_america_safe_species
```

agrupamiento por pais animales en peligro

```python
number_of_threatened_species_by_country = south_america_threatened_animals.groupby('country_name')['internalTaxonId'].count().reset_index().rename(columns={'internalTaxonId': 'Cantidad'})
number_of_threatened_species_by_country['in_risk'] = 'true'
number_of_threatened_species_by_country
```

merge de animales y su tipo de riesgo

```python
south_america_species_risk_type_count = pd.concat([number_of_south_america_safe_species, number_of_threatened_species_by_country], axis=0, ignore_index=True)
south_america_species_risk_type_count
```

```python
south_america_species_risk_type_count.to_csv(custom_datasets_path + 'south_america_species_risk_type_count.csv')
```

```python
number_of_threatened_species_by_country.to_csv(custom_datasets_path + 'number_of_threatened_species_by_country.csv')
```

##threats proportion

```python
threats_categories = latam['threats'][['internalTaxonId', 'name']]
threats_categories = threats_categories.rename(columns={'name': 'threat_category'})
threats_categories['threat_category'].unique()
```

cleaning unspecified species aka no data about threat

```python
south_america_threatened_animals_threat_categories = pd.merge(south_america_threatened_animals, threats_categories, on='internalTaxonId', how='inner')
south_america_threatened_animals_threat_categories = south_america_threatened_animals_threat_categories[:][~(south_america_threatened_animals_threat_categories['threat_category'].isin(['Unspecified species', 'Type Unknown/Unrecorded', 'Motivation Unknown/Unrecorded', 'Scale Unknown/Unrecorded', 'Trend Unknown/Unrecorded', 'Named species']))]
```

agrupamiento

```python
south_america_threatened_animals_threat_categories_count = south_america_threatened_animals_threat_categories.groupby('threat_category').count().reset_index()[['threat_category', 'internalTaxonId']]
south_america_threatened_animals_threat_categories_count = south_america_threatened_animals_threat_categories_count.rename(columns={'internalTaxonId' : 'threat_category_count'})
south_america_threatened_animals_threat_categories_count
```

```python
south_america_threatened_animals_threat_categories_count.to_csv(custom_datasets_path + 'south_america_threatened_animals_threat_categories_count.csv')
```

##Threats Categories

merge de animales en peligro con sus categories de amenaza

```python
south_america_threatened_animals_with_threats_categories = pd.merge(south_america_threatened_animals, latam['threats'][['internalTaxonId','stressName']], on='internalTaxonId', how='inner')
```

obtencion de categories por separado

```python
south_america_threats_categories_of_threathened_animals =  pd.DataFrame({'threats_categories' : south_america_threatened_animals_with_threats_categories['stressName'].str.split('|').explode()}).reset_index()
south_america_threats_categories_of_threathened_animals
```

agrupamiento y conteo de las categories

```python
south_america_threats_categories_of_threathened_animals_count = south_america_threats_categories_of_threathened_animals.groupby('threats_categories').count().rename(columns={'index': 'count'}).reset_index()
south_america_threats_categories_of_threathened_animals_count
```

```python
south_america_threats_categories_of_threathened_animals_count.to_csv(custom_datasets_path + 'south_america_threats_categories_of_threathened_animals_count.csv')
```

##Conservation Measures

```python
south_america_conservation_measures = pd.merge(south_america, latam['conservation_needed'][['internalTaxonId', 'name']], on='internalTaxonId', how='inner')
south_america_conservation_measures
```

```python
south_america_conservation_measures_count = south_america_conservation_measures.groupby('name').count().reset_index()
south_america_conservation_measures_count = south_america_conservation_measures_count.rename(columns={'name' : 'preservation_measure' , 'assessmentId' : "count"})[['preservation_measure', 'count']]
south_america_conservation_measures_count
```
